```python
# Menggunakan kolom 'movie_id' untuk clustering
X = df[['movie_id']]

# Inisialisasi dan latih model K-Means dengan jumlah kluster yang diinginkan
n_clusters = 3
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
kmeans.fit(X)

# Mendapatkan label cluster untuk setiap data
labels = kmeans.labels_

# Menambahkan kolom 'cluster' ke dalam DataFrame
df['cluster'] = labels

# Visualisasi hasil clustering
plt.figure(figsize=(10, 6))
for cluster in range(n_clusters):
    cluster_data = df[df['cluster'] == cluster]
    plt.scatter(cluster_data['movie_id'], cluster_data['title'], label=f'Cluster {cluster}')
plt.xlabel('Movie ID')
plt.ylabel('Movie Title')
plt.title('K-Means Clustering of Movie ID')
plt.legend()
plt.grid(True)
plt.show()
```

# Alasan Pemilihan Model

Dalam konteks clustering data film, terdapat beberapa pertimbangan yang menjadi alasan pemilihan model clustering tertentu. Dalam contoh ini, kita akan fokus pada pemilihan model K-Means Clustering dan Hierarchical Clustering serta penjelasan mengapa model-model tersebut dipilih:

1. K-Means Clustering
Kemudahan Implementasi: Salah satu alasan utama pemilihan K-Means Clustering adalah kemudahan implementasinya. Model ini relatif mudah dipahami dan diimplementasikan, terutama bagi mereka yang baru mengenal teknik clustering.
Skalabilitas: K-Means cocok digunakan untuk dataset yang besar karena kompleksitas komputasinya yang rendah, yang membuatnya lebih cepat dalam mengelompokkan data dibandingkan beberapa model clustering lainnya.
Ketetapan Jumlah Kluster: K-Means memerlukan penentuan jumlah kluster sebelumnya, yang cocok dalam kasus ini karena kita mungkin ingin mengelompokkan film-film ke dalam beberapa kelompok yang lebih terdefinisi.
2. Hierarchical Clustering
Pengelompokan Berhirarki: Hierarchical Clustering memungkinkan pengelompokan berhirarki, yang dapat memberikan pemahaman yang lebih dalam tentang struktur data. Ini bisa bermanfaat untuk melihat hubungan antar kelompok film secara hierarkis.
Tidak Memerlukan Jumlah Kluster Terdefinisi: Salah satu kelebihan Hierarchical Clustering adalah bahwa Anda tidak perlu menentukan jumlah kluster sebelumnya. Model ini secara otomatis membentuk hierarki kluster berdasarkan jarak antar data.
Interpretasi Visual dengan Dendrogram: Dendrogram yang dihasilkan dari Hierarchical Clustering memberikan visualisasi yang baik tentang hubungan antar kluster, yang dapat membantu dalam interpretasi hasil clustering.
Kedua model tersebut dipilih berdasarkan kebutuhan dan karakteristik data film yang digunakan dalam analisis. K-Means dipilih karena kemudahan implementasinya dan ketetapan jumlah kluster yang sesuai dengan tujuan analisis. Sementara itu, Hierarchical Clustering dipilih karena kemampuannya dalam memberikan pemahaman berhirarki tentang hubungan antar kelompok data. Kedua model ini dapat memberikan wawasan yang berharga untuk pengambilan keputusan di industri film.
